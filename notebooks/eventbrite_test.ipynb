```python
from eventbrite import Eventbrite
eventbrite = Eventbrite('S4MJIEICDYSJUQBUJTDS')
```

```python
# Get a raw list of events (includes pagination details)
events = eventbrite.event_search(**{'categories':'103','expand':['venue','ticket_availability']})
n_sold_out = 0
for p in range(events['pagination']['page_count']):
    print(p)
    ret = eventbrite.event_search(**{'categories':'103','expand':['venue','ticket_availability'], 'page':p})
    for event in ret['events']:      
        if event['ticket_availability']['is_sold_out']:
            print('got one. At',p)
            n_sold_out += 1
n_sold_out
```
